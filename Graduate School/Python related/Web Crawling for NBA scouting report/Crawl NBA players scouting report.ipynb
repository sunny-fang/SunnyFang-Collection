```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import datetime
import dateutil.parser as parser
from tqdm import tqdm
from time import sleep
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotVisibleException
```

```python
draft_data = pd.read_csv('nba_draft_combine_all_years.csv')
playerlist = draft_data.iloc[:,1]
yearlist = draft_data.iloc[:,2]
for player, year in tqdm(zip(playerlist,yearlist)):
    print(player,year)
```

```python
draft_date_2009 = parser.parse('25 JUN, 2009')
draft_date_2010 = parser.parse('24 JUN, 2010')
draft_date_2011 = parser.parse('24 JUN, 2011')
draft_date_2012 = parser.parse('29 JUN, 2012')
draft_date_2013 = parser.parse('27 JUN, 2013')
draft_date_2014 = parser.parse('29 JUN, 2014')
draft_date_2015 = parser.parse('26 JUN, 2015')
draft_date_2016 = parser.parse('23 JUN, 2016')
draft_date_2017 = parser.parse('22 JUN, 2017')
dictionary = {'2009': draft_date_2009,'2010': draft_date_2010,'2011': draft_date_2011,'2012': draft_date_2012,'2013': draft_date_2013,
             '2014': draft_date_2014,'2015': draft_date_2015,'2016': draft_date_2016,'2017': draft_date_2017,}
```

```python
# before you run this func, the web should already open
def get_player_reports(playername, year):
    driver.find_element_by_name('search').send_keys(playername)
    sleep(3)
    driver.find_element_by_xpath("//*[@class='player ui-menu-item']").click()
    testsoup = BeautifulSoup(driver.page_source, "html.parser")
    aa = testsoup.findAll('div',{'class':'article-details'})
    for index, row in enumerate(aa):
        player_name.append(playername)
        title.append(aa[index].find_next('a').text.strip())
        time = aa[index].find_next('div',{'class':'small-12 medium-12 large-4 column end date'}).text.strip()
        date.append(time)
        early_than_draft_day.append(parser.parse(time) < dictionary[str(year)])
        context.append(aa[index].find('div',{'class':'article-content'}).text.strip())
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(30)
driver.get('http://www.draftexpress.com/')
#cant_find=[];player_name=[];title=[];date=[];early_than_draft_day=[];context=[]
for player, year in tqdm(zip(playerlist[90:],yearlist[90:]),total=len(playerlist[90:])):
    try:
        get_player_reports(player,year)
    except (NoSuchElementException,ElementNotVisibleException) as e:
        cant_find.append(player)
        driver.find_element_by_name('search').send_keys(Keys.CONTROL,'a')
        driver.find_element_by_name('search').send_keys(Keys.BACKSPACE)
        pass

driver.close()
df = pd.DataFrame({'player_name':player_name,'title':title,'date':date,'early_than_draft_day':early_than_draft_day,'context':context})
df
```
